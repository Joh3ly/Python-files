## Análisis de datos de felicidad en países del mundo


Utilizando los datos sobre la felicidad publicados en español analizaremos:
Medida de tendencia central y dispersión de la felicidad a nivel mundial para el primer y último año incluido en el set de datos. 
Comparación de valores a nivel mundial con dos países en el primer año reportado y el último año reportado.
Se desea conocer cómo se comparan los países elegidos con el resto del mundo. Básicamente si son más o menos felices.


```python
#Cargamos los datos e importamos pandas
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-08-07/felicidad.csv')
```

```python
#visualizamos las variables
df.info()
```

Puntaje de felicidad, también conocido como Escalera de Vida. Indica percepción de calidad de vida en una escala de 0 a 10. Mientras más alto el puntaje, mayor felicidad. Por lo que nos enfocaremos en esta variable a analizar

```python
#Observamos los primeros 10 datos de nuesto dt
df.head(10)
```

```python
df.pais.unique()
```

Observamos todos los países que incluye nuestro dt, así como los años de referencia

```python
df.anio.describe()
```

Queremos analizar el primer y último año por lo tanto filtraremos los años 2005 que es el mínimo y 2018 que es el máximo.

```python
anio2005=df[df.anio==2005]
anio2005
```

```python
#Encontramos el promedio
anio2005.escalera_vida.mean()
```

```python
#Encontramos la varianza
anio2005.escalera_vida.std()
```

Ahora encontraremos los mismos datos para el valor máximo 2018

```python
anio2018=df[df.anio==2018]
anio2018
```

```python
#Encontramos el promedio
anio2018.escalera_vida.mean()
```

```python
#Encontramos la varianza
anio2018.escalera_vida.std()
```

Elegimos Dinamarca como primer país, podemos observar que en el 2005 el promedio de felicidad está por encima de la media no solo la media anual del 2005 sino del promedio total también. Sin embargo ha tenido una pequeña baja en el 2018, aunque sigue por sobre el promedio. Hemos considerado que quizá se deba a su expectativa de vida y baja corrupción por lo que se ha analizado estos datos. Dinamarca cuenta con una expectativa de vida de 69 años en el 2005 y 72 años en el 2018 . La varianza es mínima y en comparación con el resto del mundo es una expectativa de vida muy alta y una baja taza de percepción de corrupción. Y aunque la felicidad ha bajado apenas un punto, vemos que ha aumentado su expectativa de vida y bajado la percepción de corrupción por lo que creemos que estas variables si guardan relación.

Elegimos Japón como segundo país, podemos observar que en el 2005 el promedio de felicidad está por encima de la media aunque no sea tan alto ni se haya mantenido sigue por encima del promedio en el 2005. Para el 2018 está apenas en el promedio, sin embargo no se observó variabilidad en la percepción de corrupción, pero sí un aumento de la expectativa de vida que está casi 11 años por encima del promedio en 2018.

Si tomamos en consideración que países han diso más felices tomando en cuenta las variables, expectativa de vida, escalera de vida y percepcion de corrupción de todos los países desde el 2005 hasta el 2018.

```python
prueba=df.pivot_table(index='pais', values=['escalera_vida', 'expectativa_vida','percepcion_corrupcion'], aggfunc=['mean'])
prueba
```

Hemos creado una tabla con los promedios por país de las variables de interés. Ahora veremos el top 10 de países con los max de cada variable.

```python
prueba.sort_values(by=[('mean','escalera_vida')], ascending=False).head(10)
```

```python
prueba.sort_values(by=[('mean','expectativa_vida')], ascending=False).head(10)
```

```python
prueba.sort_values(by=[('mean','percepcion_corrupcion')]).head(10)
```

```python
prueba.sort_values(by=[('mean','escalera_vida'),('mean','expectativa_vida')], ascending=False).head(10)
```
